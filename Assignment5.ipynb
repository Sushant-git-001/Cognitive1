```python
#1
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
total_sum= np.sum(gfg)
column_sum=np.sum(gfg,axis=0)
row_sum=np.sum(gfg,axis=1)

print(total_sum)
print(column_sum)
print(row_sum)

```

```python

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
int_array=array[array==np.floor(array)]
print(int_array)
float_array=array[array!=np.floor(array)]
print(float_array)
```

```python
#4
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-10,10,100)
y_square=x**2
y_sin=np.sin(x)
y_exp=np.exp(x)
y_log=np.log(np.abs(x)+1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('functions graph')
plt.plot(x,y_square,label='y=x^2',linestyle='--',color='r')
plt.plot(x,y_sin,label='y=sin(x)',linestyle='-',color='b')
plt.plot(x,y_exp,label='y=exp(x)',linestyle=':',color='g')
plt.plot(x,y_log,label='y=log(x)',linestyle='-.',color='m')
plt.legend()
plt.show()
```
