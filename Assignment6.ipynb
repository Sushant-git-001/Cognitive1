```python
#1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
M=int(input("Enter the Value M for Function: "));
x=np.linspace(-10,10,300);
y1=M*(x**2);
y2=M*(np.sin(x));
plt.plot(x,y1,label=f'y1={M}*x^2',color='r',linestyle='dashed');
plt.plot(x,y2,label=f'y2={M}*sin(x)',color='b',linestyle='dotted');
plt.xlabel('x');
plt.ylabel('y');
plt.title('Graph of y1 and y2');
plt.legend();
plt.grid();
plt.show();




```

```python
#2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Score': [65, 90, 95, 58, 75]
}

df = pd.DataFrame(data);

sns.set_style("whitegrid");

plt.figure(figsize=(8, 5));
barplot = sns.barplot(x='Subject', y='Score', data=df, palette="viridis");

for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Student Scores", fontsize=14);
plt.xlabel("Subjects", fontsize=12);
plt.ylabel("Scores", fontsize=12);
plt.ylim(0, 100);
plt.grid(axis='y', linestyle='--');

plt.show();
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(102317141)

data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), marker='o', color='b', linestyle='dashed')
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

x_values = np.arange(50)
y_values = data + np.random.normal(0, 0.2, 50)
axes[0, 1].scatter(x_values, y_values, color='r', marker='o')
axes[0, 1].set_title("Scatter Plot with Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Values")
axes[0, 1].grid(True)

axes[1, 0].axis("off")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
#4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "C:\Users\susha\Downloads\company_sales_data.csv"
df = pd.read_csv(file_path)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 4))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker="o", color="b")
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.show()


plt.figure(figsize=(8, 4))
for col in ["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]:
    sns.lineplot(x=df["month_number"], y=df[col], marker="o", label=col)
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

df.set_index("month_number").plot(kind="bar", figsize=(10, 5))
plt.title("Bar Chart for All Features")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend(title="Attributes", bbox_to_anchor=(1, 1))
plt.show()
```
